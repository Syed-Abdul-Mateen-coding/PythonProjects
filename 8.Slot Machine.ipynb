```python
import random as r

def spin(lines,per_line):
    v=["A","B","C","D"]
    p={"A":5,"B":4,"C":3,"D":2}

    
    roll = r.choices(v,weights=(2,4,6,8),k=9)
    matched_line={}
    
    
    for index,val in enumerate(roll,start=1):
        
        if(index%3!=0):
            print(val+" |",end=" ")
        else:
            print(val)
            if (len(list(set(roll[index-3:index])))==1):
                matched_line[index//3]=roll[index-1]


    win=0
    
    if(matched_line):
        print(f"{len(matched_line)} line(s) has matched")

        for i in (list(matched_line.keys())):
            if(i in range(1,lines+1)):
                win+=p[matched_line[i]]*per_line

            
        
    if(not(matched_line)):
      print("Oh shit! no line has matched")

    print(f"lines matched: {list(matched_line.items())}")

    return win 

           

def bet(depo):
    print(f"your balance is: {depo}")
    l=int(input("How many lines do you want to bet on (1-3): "))
    while(l not in range(1,4)):
        l=int(input("please enter valid line(s) you want to bet on: "))

    m=int(input("How much money do you want to bet on each line: "))
    while(m not in range(10,101)):
        m=int(input("please enter valid amount you want to bet on each line (10-100): "))
        
    if(depo<m*l):
        print("You donot have enough balance to bet")
        add=input("do you want to add amount in the pool (y/n): ")
        while(add.lower() not in ("y","n")):
            add=int(input("please enter valid action: "))
                    
        if add.lower()=="y":
            a=int(input(f"Enter the amount you want to add in the pool of {depo}$ (Minimum amount 10$): "))
            while(a<10):
                a=int(input("please enter valid amount you want to bet on (10-100): "))
            print(f"your balance is now {depo+a}")
            print()
            bet(a+depo)
                    
        else:
            e=input("Enter 1 to end the bet or 2 to play again : ")
            while(e.lower() not in ("1","2")):
                e=input("please enter valid action (1,2): ")
            if(e=="1"):
                return
            else:
                bet(depo)
                        

    else:
        balance=depo-(m*l)
        print(f"your balance is now: {balance}")
        
        s=spin(l,m)
        
        if(s>0):
            print(f"Hurray! you won {s}$ from this bet")
            print(f"Great! now your balance is {s+balance}$")

        else:
            print(f"Oh no! now your balance is now {s+balance}$")

        print()

        a=input("Do you want to play again (y/n): ")
        while(a.lower() not in ("y","n")):
            a=input("please enter valid action (y,n): ")

        if a.lower()=="y":
            print()
            bet(balance+s)
        else:
            return


s=int(input("How much do you want to depsit (Mininmum deposit is 10$): "))
while(s<=10):
    s=input("Please deposit a valid ammount to bet in ($): ")


bet(s)

print()
print("Thankyou for betting at our slot!")
```
